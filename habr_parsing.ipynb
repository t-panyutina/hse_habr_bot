```python
import requests 
from bs4 import BeautifulSoup

import pandas as pd
import csv
```

```python
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.chrome.options import Options
```

```python
def run_browser(): 
    global driver
    options = Options()
    options.headless = True
    driver = webdriver.Chrome(options=options)
    driver.set_page_load_timeout(100)
```

```python
def get_pages(n: int, category: str) -> list:
    pages = []
    for i in range(1, n):
        pages.append(f'https://habr.com/ru/hubs/{category}/articles/page{i}/')
    return pages
```

```python
def get_page_links(pages: list) -> list:
    links = []
    run_browser()
    for page in pages:
        try:
            driver.get(page)
            page1 = driver.page_source
            soup1 = BeautifulSoup(page1, 'lxml')
            links += ['https://habr.com' + el['href'] for el in soup1.find_all("a", class_="tm-title__link")]
        except TimeoutException as ex:
            print("Exception has been thrown. " + str(ex)) 
    driver.quit()
    
    return links
```

```python
df = pd.DataFrame(columns=['category', 'link', 'text', 'title'])
```

```python
def get_article(link: str, category: str) -> dict:
    article_dict = {}
    try:
        driver.get(link)
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'lxml')
        article_dict['category'] = category
        article_dict['link'] = link
        if soup.find('div', attrs={'id': 'post-content-body'}) is None:
            article_dict['text'] = 'Без текста'
        else:
            article_dict['text'] = soup.find('div', attrs={'id': 'post-content-body'}).text
        if soup.find('h1', class_='tm-title tm-title_h1') is None:
            article_dict['title'] = 'Без названия'
        else:
            article_dict['title'] = soup.find('h1', class_='tm-title tm-title_h1').text.replace("'", "")
    except TimeoutException as ex:
        print("Exception has been thrown. " + str(ex))
    
    return article_dict
```

```python
devops_pages = get_pages(50, 'devops')
```

```python
devops_links = get_page_links(devops_pages)
```

```python
for number, link in enumerate(devops_links):
    if (number/30).is_integer():
        driver.quit()
        run_browser()
        df = pd.concat([df, pd.DataFrame([get_article(link, 'devops')])], ignore_index=True)
    else:
        df = pd.concat([df, pd.DataFrame([get_article(link, 'devops')])], ignore_index=True)
```
